#***1). Import the libraries :-***

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

#***2). Load the image :-***

```python
image = cv2.imread("Usain.Bolt.jpg")
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

#***3). Converting Image to Gray Scale :-***

```python
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

#***4). Inverting the Gray Image :-***

```python
image_invert = cv2.bitwise_not(img_gray)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(image_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

#***5). Blur the image:-***

```python
blur_image=cv2.GaussianBlur(image_invert,(21,21),sigmaX=0, sigmaY=0)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(blur_image,cmap="gray")
plt.axis("off")
plt.title("Blurry Image")
plt.show()
```

#***6). Here is your Final Pencil sketch***

```python
sketch = cv2.divide(img_gray, 255 - blur_image, scale=255)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(sketch,cmap="gray")
plt.axis("off")
plt.title("Final Sketch")
plt.show()
```
